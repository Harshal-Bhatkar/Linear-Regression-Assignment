## Step5: Data Preperation for Linear Regression

```python
data.head()
```

## Step6: Model Building and Trainning Data

```python

#Split Data into training and test

from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively

df_train, df_test = train_test_split(data, train_size = 0.70, random_state = 100)
```

```python
df_train.head()
```

```python
# check the shape of training datatset

df_train.shape
```

```python
# check the shape of test datatset

df_test.shape
```

```python
#Rescaling the features Using MinMax Scaling

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables

num_vars = ['temp','atemp','hum','windspeed','cnt']
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

```python
# verifying the head after appying scaling.

df_train.head()
```

```python

df_train.describe()
```

```python
## creating the X anf y variable
y_train = df_train.pop('cnt')
X_train = df_train
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
##Analyzing the Corelation matrix among the variables
plt.figure(figsize=(30,25))
sns.heatmap(data.corr() , annot = True, cmap='RdYlBu')
plt.show()
```

#### Note: 
- We have almost 30 Predictors to Predict our outcome. Analysing these many predictors can be very hectic.
- To simplify this process, we will use RFE (Recursive Feature Elimination) method which will provide us with set of variables ranked in an order of importance.
- We will pass an arbitary number (here 15), which means the method will rank first 15 predictors as 1. Creating a generalised function for evaluating Variance Inflation Factor.

```python
## Using RFE method

# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# RFE method for feature Elimination

lm = LinearRegression()
lm.fit(X_train,y_train)

rfe = RFE(lm,15)
rfe.fit(X_train,y_train)
```

```python
rfe_df = pd.DataFrame({'Predictor': X_train.columns, 'Select Status': rfe.support_, 'Ranking': rfe.ranking_})
rfe_df.sort_values(by='Ranking')
```

```python
# selecting the RFE columns
column = X_train.columns[rfe.support_]
column
```

##### Building the Linear Model

```python
## Importing Libraries
import statsmodels.api as sm 
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python

X_train_RFE = X_train[column]
```

```python
## Checking for relation between the seleted variables

plt.figure(figsize=(15,8))
sns.heatmap(X_train_RFE.corr(),annot = True, cmap="RdYlGn",linewidth =1)
plt.show()
```

####  Since there is a high correlation between ( weekday_6 and workingday) and also between (temp and season_spring) we might have to delete one variable has it will create issues in the model first will check P values and VIF and based on that we will try to take descion

```python
## Building First Model
X_train_lm_1 = sm.add_constant(X_train_RFE)
lr_1 = sm.OLS(y_train,X_train_lm_1).fit()
print(lr_1.summary())
```

```python
vif = pd.DataFrame()
X = X_train_RFE
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# As humidity shows very high VIF values hence we can drop it
X_train_RFE = X_train_RFE.drop(['hum'], axis = 1)
```

```python
## Building Second Model
X_train_lm_2 = sm.add_constant(X_train_RFE)
lr_2 = sm.OLS(y_train,X_train_lm_2).fit()
print(lr_2.summary())
```

```python
vif = pd.DataFrame()
X = X_train_RFE
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# As vif valur of temp is higher but we cannot delete this variable because of high corelation with cnt we will go with the manual approac
# As there is high correlation between season_spring and temp we can keep only one data so will remove season_spring as by looking at the relation with the target variable count
X_train_RFE = X_train_RFE.drop(['season_spring'], axis = 1)
```

```python
## Building third Model
X_train_lm_3 = sm.add_constant(X_train_RFE)
lr_3 = sm.OLS(y_train,X_train_lm_3).fit()
print(lr_3.summary())
```

```python
vif = pd.DataFrame()
X = X_train_RFE
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# As vif valur of temp is higher but we cannot delete this variable because of high corelation with cnt we will go with the manual approac
# As there is high correlation between weekday_6 and workingday we can keep only one data so will remove weekday_6 as by looking at the relation with the target variable count
X_train_RFE = X_train_RFE.drop(['weekday_6'], axis = 1)
```

```python
## Building fourth Model
X_train_lm_4 = sm.add_constant(X_train_RFE)
lr_4 = sm.OLS(y_train,X_train_lm_4).fit()
print(lr_4.summary())
```

```python
vif = pd.DataFrame()
X = X_train_RFE
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# As workingday shows very high VIF values hence we can drop it
X_train_RFE = X_train_RFE.drop(['workingday'], axis = 1)
```

```python
## Building fifth Model
X_train_lm_5 = sm.add_constant(X_train_RFE)
lr_5 = sm.OLS(y_train,X_train_lm_5).fit()
print(lr_5.summary())
```

```python
vif = pd.DataFrame()
X = X_train_RFE
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# As mnth_july is having negative coefficients and has highest P Value will drop it
X_train_RFE = X_train_RFE.drop(['mnth_july'], axis = 1)
```

```python
## Building sixth Model
X_train_lm_6 = sm.add_constant(X_train_RFE)
lr_6 = sm.OLS(y_train,X_train_lm_6).fit()
print(lr_6.summary())
```

```python
vif = pd.DataFrame()
X = X_train_RFE
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

```python
# Checking the parameters and their coefficient values
lr_6.params
```

#### Linearity of model

```python
plt.figure(figsize=(8,6))
plt.scatter(y_train, y_train_pred, c='crimson')
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
p1 = max(max(y_train_pred), max(y_train))
p2 = min(min(y_train_pred), min(y_train))
plt.plot([p1, p2], [p1, p2], 'b-')
```

#### Observation:
- The actual value and predicted value are showing linearly corelation
